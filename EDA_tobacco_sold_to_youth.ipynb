```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("tobacco-sold-to-youth.csv")
display(df)
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()

if df.isnull().sum().sum() == 0:
  print("there is no missing value")
else:
  print("there is missing value")
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x=df['YEAR'], y=df['PERCENT'], marker='o', label='Percent Sold')
#plt.fill_between(df['YEAR'], df['LOWER95'], df['UPPER95'], alpha=0.3, label='95% Confidence Interval')
plt.xlabel('Year')
plt.ylabel('Percentage Sold')
plt.title('Trend of Tobacco Sold to Youth Over the Years')
plt.grid()
plt.legend()
plt.show()
```
