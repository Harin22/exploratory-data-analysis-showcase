# Exploratory Data Analysis (EDA) of tobacco sales to youth

*objective was to analyze historical data, identify patterns, detect anomalies, and provide insights into how tobacco sales have changed over the years.*

*   Identify trends
*   List item



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("tobacco-sold-to-youth.csv")
display(df)
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()

if df.isnull().sum().sum() == 0:
  print("there is no missing value")
else:
  print("there is missing value")
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x=df['YEAR'], y=df['PERCENT'], label='Percent Sold')
plt.xlabel('year')
plt.ylabel('Percentage sold')
plt.title('Trend of tobacco sold to youth Over years')
plt.grid()
plt.legend()
plt.show()
```

```python
last_10_year = df.tail(10)

plt.figure(figsize = (10,5))
plt.grid()
sns.barplot(x=last_10_year['YEAR'], y=last_10_year['PERCENT'], label='Percent Sold')
```

```python
particular_year = last_10_year[last_10_year['YEAR'] == 2018]
display(particular_year)
print("\n ")

df[df['YEAR'].isin([2016, 2017, 2018, 2019])]
```

```python
last_20_years = df.tail(20)
percentage_decreased = last_20_years["PERCENT"].min()
percentage_increased = last_20_years["PERCENT"].max()
Decline_over_the_year_1 = percentage_increased - percentage_decreased
Decline_over_the_year_2 = Decline_over_the_year_1 / percentage_increased
final = Decline_over_the_year_2 * 100
display(final)
rounded_final = round(final, 2)
print("the tobaco has been decreased over the year : ", rounded_final," %")

```
